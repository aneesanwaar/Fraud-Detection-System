**1. Data Preprocessing:**

Loading Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#loading the dataset
df = pd.read_csv("/content/creditcard.csv")
```

```python
print(df.info())
```

```python
#checking for missing values
print(df.isnull().sum())
```

```python
# removing all rows with missing values to avoid introducing bias
df = df.dropna()
```

```python
#checking again for missing values
print(df.isnull().sum())
```

```python
#checking the class distribution
print(df["Class"].value_counts())
```

So this Dataset is highly unblanced

Applying random undersampling to the majority class (Class = 0) to match the minority class (Class = 1):

```python
from imblearn.under_sampling import RandomUnderSampler
```

```python
# defining undersampling strategy
undersampler = RandomUnderSampler(random_state=42)
```

```python
# spliting features (X) and target (y)
X = df.drop(columns=['Class'])
y = df['Class']
```

```python
# applying undersampling
X_resampled, y_resampled = undersampler.fit_resample(X, y)
```

```python
# checking new class distribution
print(pd.Series(y_resampled).value_counts())
```

Spliting data into Training and Testing sets

```python
from sklearn.model_selection import train_test_split
```

```python
# splitting the balanced dataset into training (80%) and testing (20%) sets
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=42, stratify=y_resampled)
```

```python
from sklearn.preprocessing import StandardScaler

# Initialize the scaler
scaler = StandardScaler()

# Fit on the training data and transform both train and test sets
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

**2. Model Training:**

Training a Random Forest Model

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
```

```python
# Initializing Random Forest
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
```

```python
#training random forest
rf_model.fit(X_train, y_train)
```

```python
# Making predictions
y_pred = rf_model.predict(X_test)
```

**3. Model Evaluation:**

Evaluating the system's precision, recall, and F1-score.

```python
# evaluation metrics
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```
